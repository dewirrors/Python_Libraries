```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor
```

```python
#Task_1

bhp_dataset = load_boston()
print(f'Набор данных "Boston House Prices dataset":\n{bhp_dataset}')
```

```python
X = pd.DataFrame(bhp_dataset['data'], columns=bhp_dataset['feature_names'])
X
```

```python
y = pd.DataFrame(bhp_dataset['target'], columns=['price'])
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

y_predict_lr = lr.predict(X_test)
print(f'Предсказание значений тестового датафрейма y моделью линейной регрессии:\n{y_predict_lr}')
```

```python
r2_lr = r2_score(y_test, y_predict_lr)
print(f'Значение коэффициента детерминации для прогноза модели линейной регрессии: {r2_lr}')
```

```python
#Task_2

model = RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=42)
model.fit(X_train, y_train.values[:, 0])

y_predict_model = model.predict(X_test)
print(f'Предсказание значений тестового датафрейма y моделью случайного леса:\n{y_predict_model}')
```

```python
r2_model = r2_score(y_test, y_predict_model)
print(f'Значение коэффициента детерминации для прогноза модели случайного леса: {r2_model}')
```

```python
#По коэффициенту детерминации можно сделать вывод о превосходстве модели случайного леса над моделью линейной регрессии,
#так как для прогноза модели случайного леса высчитан более высокий показатель коэффициента детерминации.
#Однако для более точной оценки эффективности моделей необходимо провести всесторонний анализ их качества,
#в том чсиле оценив вероятность переобучения у обеих моделей.
```
